```python
import random
#1a)
def genereaza(n):
  numbers =[]
  for i in range(1, n):
    random_number = random.randint(1, 365)
    if random_number in numbers:
      return True
    numbers = numbers + [random_number]
  return False

def testeaza(n):
  cf = 0
  ct = 0
  for i in range(10000):
    if genereaza(n):
      cf += 1
    ct += 1
    return cf / ct
```

```python
#1b)
number = input("Introduceti un numar: ")

print("Probabiliatea a fost: ")
prob = testeaza(int(number))
print("{:.2f}".format(prob*100) , "%")
```

```python
#1c
from matplotlib.pyplot import plot, grid, title, show

title('Plot probability')
xs = range(2, 50)
ys = [testeaza(x) for x in xs]
plot(xs,ys,'r*')
grid()
show()
```

```python
#2 i)
from matplotlib.pyplot import axis, plot, show
from random import random
from math import dist

bulina_albastra = 0

for i in range(1000):

  axis('square')
  axis((0, 1, 0, 1))
  x = random()
  y = random()
  F=[x,y]
  if dist(F, [0.5, 0.5]) < 0.5:
    plot(F[0],F[1],'bo')
    bulina_albastra += 1
  else:
    plot(F[0],F[1],'ro')
print("{:.2f}%".format(bulina_albastra/1000* 4 ))
show()
```

```python
#iii)

from matplotlib.pyplot import axis, plot, show
from random import random
from math import dist

bulina_albastra = 0


for i in range(2000):

  axis('square')
  axis((0, 1, 0, 1))
  x = random()
  y = random()
  F=[x,y]
  d1 = dist([1, 1], [x, y])
  d2 = dist([1, 0], [x, y])
  d3 = dist([0, 0], [x, y])
  d4 = dist([0, 1], [x, y])
  rez = 0
  if d1**2 + d2**2 < 1:
    rez -= 1
  else:
    rez +=1
  if d2**2 + d3**2 < 1:
    rez -=1
  else:
    rez +=1
  if d3**2 + d4**2 < 1:
    rez -= 1
  else:
    rez +=1
  if d4**2 + d1**2 < 1:
    rez -=1
  else:
    rez +=1
  if rez == 0:
    plot(F[0],F[1],'bo')
    bulina_albastra += 1
  else:
    plot(F[0],F[1],'ro')
print("{:.2f}%".format(bulina_albastra/1000* 4 ))
show()
```
