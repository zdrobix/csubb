```python
#problema 1i)
import random

def p1():
  bile = []
  posibilitati = ['r', 'r', 'r', 'r', 'r', 'a', 'a', 'a', 'v', 'v']
  while (len(bile) < 4):
    bile.append(random.choice(posibilitati))

  toate_la_fel = all(x == bile[0] for x in bile)

  una_rosie = False
  for b in bile:
    if b == 'r' and toate_la_fel == False:
      return 'unarosie'

  if toate_la_fel:
    return 'toatelafel'

  return None
```

```python
#problema 1i)
def test_p1():
  una_rosie =0
  toate_la_fel =0
  for i in range(0, 100000):
    if p1() == 'unarosie':
      una_rosie += 1
    if p1() == 'toatelafel':
      toate_la_fel += 1
  prob = toate_la_fel / una_rosie
  print("{:.2f}".format(prob*100) , "%")

test_p1()
```

```python
#problema 1ii)
#probabilitatea teoretica
# P(B|A) = countA_B / countA = (countA_B / nr_sim) / (countA / nr_sim)
#                            = P(A intersectat B) / P(A)
#formula probabilitatii conditionate

#P(A) = 1 - P(!A) = 1 - C de 5 cate 3 / C de 10 cate 3
# (comb(5, 3) / comb(10, 3) / (1 - comb(5, 3) / comb(10, 3) )
#P(A intersectat B) = C de 5 cate 3 / C de 10 cate 3

#A: cel putin o bila rosie:
#!A : niciuna rosie

from math import comb

prob = (comb(5, 3) / comb(10, 3) / (1 - comb(5, 3) / comb(10, 3) ))
print("{:.2f}".format(prob*100) , "%")
```

```python
#problema 2)
from random import randrange
from matplotlib.pyplot import bar, hist, grid, show, legend

data = [randrange(1,7) for _ in range(50000)]
bin_edges = [k+0.5 for k in range(7)]
hist(data, bin_edges, density = True, rwidth = 0.9, color = 'green', edgecolor = 'black',
alpha = 0.5, label = 'frecvente relative')
distribution = dict([(i,1/6) for i in range(1,7)])
bar(distribution.keys(), distribution.values(), width = 0.85, color = 'red', edgecolor = 'black',
alpha= 0.6, label = 'probabilitati')
legend(loc = 'lower left')
grid()
show()
```
