```python
from scipy.stats import uniform
from matplotlib.pyplot import bar, show, hist, grid, legend, xticks, yticks, plot
import numpy as np
from math import log
from scipy.stats import expon
```

```python
#1)
def genereaza (lista_numere, lista_probabilitati):
  x = uniform.rvs()
  count = -1
  for p in lista_probabilitati:
    count += 1
    x -= lista_probabilitati[count]
    if x < 0:
      return lista_numere[count]


```

```python
[genereaza([1, 2, 3], [8/10, 1/10, 1/10]) for _ in range(10)]
```

```python
grupe_sangv = ['0', 'A', 'B', 'AB']
prob = [46/100, 40/100, 1/10, 4/100]
data = [genereaza(grupe_sangv, prob) for _ in range(1000)]

bin_edges = np.arange(len(grupe_sangv) + 1)
hist(data, bin_edges, density = True, rwidth = 0.9, color = 'green', edgecolor = 'black', alpha = 0.5, label = 'frecvente relative')
distribution = dict([(i,1/4) for i in range(1,5)])
legend(loc = 'lower left')
grid()
show()
```

```python
#2
alpha = 12
data = expon.rvs(loc = 0, scale = alpha, size = 1000)
hist(data, bins = 12, density = True, range=(0, 60))
x = range(60)
plot(x, expon.pdf(x, loc= 0, scale = 12), 'r-')

xticks(range(0, 60, 5))
grid()
show()
```
