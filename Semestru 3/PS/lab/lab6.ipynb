```python
from scipy.stats import norm, bernoulli
from numpy import mean, std, linspace
from matplotlib.pyplot import show, hist, grid, legend, xticks, plot
from scipy.stats import expon, uniform
from numpy import mean, std, multiply
from scipy.stats import uniform
from numpy import exp, mean
from scipy.integrate import quad
```

```python
#1
def generate (x, n):
  return norm.rvs(loc = x, scale = 10, size = n)

```

```python
#i)
data = generate(165, 1000)
hist(data, bins = 14, density = True, range=(130, 200))
x = range(130, 200)
plot(x, norm.pdf(x, loc= 165, scale = 10), 'r-')

xticks(range(130, 205, 5))
grid()
show()
```

```python
#ii)

data = generate(165, 1000)
vmed = sum([160 < x < 170 for x in data]) / len(data)
print("Probabilitea este ", vmed)

print("Valoarea medie", data.mean())

print("Deviatia standard", data.std())
```

```python
#2. a)
#i1 => 0.4  -> t1 (exp(1/5))
#i2 => 0.6  -> t2(unif[4, 6])
x = bernoulli.rvs(p = 0.4, size = 1000)
data = [imp * expon.rvs(loc = 0, scale = 5) + (1 - imp) * uniform.rvs(loc = 4, scale = 2) for imp in x]
print("Valoarea medie ", mean(data))
print("Deviatia standard ", std(data))

#2. b)
data_sub5 = [ x < 5 for x in data]
print("Probabilitatea ca timpul sa fie sub 5 sec ", sum(data_sub5) / len(data))
```

```python
#2. c)
prob = 0.4 * expon.cdf(5, scale = 5) + 0.6 * uniform.cdf(5, loc = 4, scale = 2)
print(prob)
```
