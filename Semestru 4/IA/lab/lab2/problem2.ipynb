```python
import os
from PIL import Image, ImageFilter
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
img = Image.open('./images/Norvig.jpg').convert('L')

plt.imshow(img, cmap='gray')  
plt.axis('off') 
plt.show()
```

```python
images = []

for filename in os.listdir('./images'):
    img = Image.open('./images/' + filename).convert('L')
    img = img.resize((128, 128), Image.Resampling.LANCZOS)
    images += [img]
    
fig, axes = plt.subplots(4, 3, figsize=(8.0, 6.0))
axes = axes.flatten()

for i in range(len(images)):
    axes[i].imshow(images[i], cmap='gray')
    axes[i].axis('off')
    
for i in range(len(images), len(axes)):
    axes[i].axis('off')
    
plt.tight_layout()
plt.show()
```

```python
img = Image.open('./images/Norvig.jpg').convert('L')
img_blurred = img.filter(ImageFilter.GaussianBlur(radius=3))

img_array = np.array(img)
img_blurred_array = np.array(img_blurred)

fix, axes = plt.subplots(1, 2, figsize=(5.0, 10.0))
axes = axes.flatten()

images = [255 - img_array, 255 - img_blurred_array]
titles = ['Before', 'After']

for i in range(2):
    axes[i].imshow(images[i], cmap='gray', vmin=0, vmax=255)
    axes[i].axis('off')
    axes[i].set_title(titles[i], fontsize=10)
```

```python
img = Image.open('./images/Norvig.jpg').convert('L')
img_edges = cv2.imread('./images/Norvig.jpg', cv2.IMREAD_GRAYSCALE)

edges = cv2.Canny(img_edges, 100, 200)

fix, axes = plt.subplots(1, 2, figsize=(5.0, 10.0))
axes = axes.flatten()

axes[0].imshow(255 - np.array(img), cmap='gray', vmin=0, vmax=255)
axes[0].axis('off')
axes[0].set_title('Before', fontsize=10)

axes[1].imshow(255 - np.array(edges), cmap='gray')
axes[1].axis('off')
axes[1].set_title('After', fontsize=10)
plt.show()
```
